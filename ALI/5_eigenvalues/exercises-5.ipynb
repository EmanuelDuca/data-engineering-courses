```python
# Exercise 6 (5.1.40)
s = symbols('s')
B = Matrix([
    [(-23-s), 57,  -9, -15, -59],
    [-10, (12-s), -10,   2, -22],
    [ 11,  5,  (-3-s), -19, -15],
    [-27, 31, -27,  (25-s), -37],
    [ -5,-15,  -5,   1,  (31-s)]
])
display(B)

determinant = B.det()
display(determinant)

roots = solve(determinant, s)
display(roots)

print("Test first solution alptha = -14")
alptha_1 = -14

B = Matrix([
    [(-23-alptha_1), 57,  -9, -15, -59],
    [-10, (12-alptha_1), -10,   2, -22],
    [ 11,  5,  (-3-alptha_1), -19, -15],
    [-27, 31, -27,  (25-alptha_1), -37],
    [ -5,-15,  -5,   1,  (31-alptha_1)]
])
display(B.rref())



A = Matrix([
    [-23, 57,  -9, -15, -59],
    [-10, 12, -10,   2, -22],
    [ 11,  5,  -3, -19, -15],
    [-27, 31, -27,  25, -37],
    [ -5,-15,  -5,   1,  31]
])
display(A.eigenvects())

```
