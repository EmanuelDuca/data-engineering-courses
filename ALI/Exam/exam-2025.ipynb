```python
# Exercise 1
A = Matrix([[0, -2], [-1, -2]])
B = Matrix([[1, 0], [1, 1]])
C = Matrix([[1, 1], [1, 2]])

display(A.det())
display(B.det())
display(C.det())

Adet = -2
Bdet = 1
Cdet = 1
BAC = A*B*C
display(BAC.det())
CAB = C*A*B
display(CAB.det())

M = Matrix([
    [Adet, Cdet, BAC.det()],
    [Cdet, -Bdet, CAB.det()],
])
display(M)
display(M.rref())
```

```python
#Exercise 2
r, s, p, q = symbols('r s p q')

V_base = Matrix([2,0,1,1])
V_r = Matrix([1,0,-1,2])
V_s = Matrix([0,1,1,-1])

W_base = Matrix([3,1,1,0])
W_p = Matrix([-1,2,1,0])
W_q = Matrix([2,-1,0,1])

eq = Eq(V_base + r*V_r + s*V_s, W_base + p*W_p + q*W_q)


eqs = []
for i in range(4):
    eqs.append(eq.lhs[i] - eq.rhs[i])


sol = solve(eqs, (r, s, p, q), dict=True)[0]


intersection = V_base + sol[r]*V_r + sol[s]*V_s
intersection = intersection.applyfunc(lambda x: x if not x.is_Rational else Rational(x))

display(intersection)
```
