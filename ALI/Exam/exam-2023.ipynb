```python
U = Matrix([[2,-1,-2]])
p = Matrix([[1, 1, 2]])

zeroMatrix = Matrix.zeros(1, 3)

# Basis vectors
u1 = Matrix([1, 2, 0])
u2 = Matrix([0, -2, 1])

# Gram-Schmidt process
def gram_schmidt(vectors):
    ortho = []
    for v in vectors:
        for u in ortho:
            v = v - (v.dot(u)/u.dot(u))*u
        ortho.append(v)
    return ortho

ortho_basis = gram_schmidt([u1, u2])
display(ortho_basis)

# Orthonormalize
e1 = ortho_basis[0] / ortho_basis[0].norm()
e2 = ortho_basis[1] / ortho_basis[1].norm()

# Vector to project
p = Matrix([1, 1, 2])

# Projection formula
proj = (p.dot(e1))*e1 + (p.dot(e2))*e2
display(proj)
```

```python
### Chat Example 
# q vector
q = Matrix([3, 1, 0])

# Subspace U: 2x - y - 2z = 0
# General solution: [x, 2x-2z, z] = a*[1,2,0] + b*[0,-2,1]
# Let u = a*[1,2,0] + b*[0,-2,1]
a, b = symbols('a b')
u = Matrix([a, 2*a - 2*b, b])

# Distance squared between q and u
diff = q - u
dist2 = diff.dot(diff)

# Minimize distance: take partial derivatives and solve
sol = solve([dist2.diff(a), dist2.diff(b)], (a, b))
a_star, b_star = sol[a], sol[b]

# Closest point in U
u_star = Matrix([a_star, 2*a_star - 2*b_star, b_star])

# Distance
distance = sqrt((q - u_star).dot(q - u_star))
distance = distance.simplify()
display(distance)

# As irreducible fraction
from sympy import fraction, nsimplify
frac = nsimplify(distance)
num, denom = fraction(frac)
display(num, denom)
```
