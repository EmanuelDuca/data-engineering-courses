```python
#An analysis/description of which questions are most crucial concerning their placement on the axes.
##Data prep
import pandas as pd
from sklearn.preprocessing import StandardScaler
alldata = pd.read_excel("alldata.xlsx")

meta_columns = ['navn', 'parti', 'storkreds', 'alder']
question_columns = [col for col in alldata.columns if col not in meta_columns]

alldata["candidate_id"] = alldata["navn"] + " (" + alldata["parti"] + ")"

pivot_df = alldata.set_index("candidate_id")[question_columns]

pivot_df = pivot_df.apply(pd.to_numeric, errors='coerce').fillna(0)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(pivot_df)

# Now X_scaled is processed

```

```python

##PCA
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt

pca = PCA(n_components=2)  # Two axes
X_pca = pca.fit_transform(X_scaled)

pc1_loadings = pd.Series(pca.components_[0], index=pivot_df.columns)
top_pc1 = pc1_loadings.abs().sort_values(ascending=False).head(10)
important_questions = pd.DataFrame({
    "question_id": top_pc1.index,
    "loading": pc1_loadings[top_pc1.index],
    "abs_loading": top_pc1.values
})

print("Top 10 questions contributing to the 1st principal component (main ideological axis):")
print(important_questions)

plt.figure(figsize=(10, 6))
plt.barh(important_questions["question_id"], important_questions["loading"], color='steelblue')
plt.axvline(0, color='black', linestyle='--')
plt.xlabel("PC1 Loading")
plt.title("Top 10 Questions Influencing Ideological Axis (PC1)")
plt.tight_layout()
plt.gca().invert_yaxis()
plt.show()
```

```python


long_df = alldata.melt(id_vars='parti', value_vars=question_columns,
                       var_name='question', value_name='response')

long_df['response'] = pd.to_numeric(long_df['response'], errors='coerce')

long_df = long_df.dropna(subset=['response'])

avg_per_party = long_df.groupby(['question', 'parti'])['response'].mean().reset_index()

pivot_avg = avg_per_party.pivot(index='question', columns='parti', values='response')

print(pivot_avg.head())
```

```python


##More data handling
import pandas as pd

drq = pd.read_excel("drq.xlsx")  
tvq = pd.read_excel("tv2q.xlsx")  

drq_qmap = drq[['ID', 'Question']].dropna()
drq_qmap.columns = ['id', 'question']
drq_qmap['id'] = drq_qmap['id'].astype(str)

tvq_qmap = tvq[['id', 'question']].dropna()
tvq_qmap['id'] = tvq_qmap['id'].astype(str)

combined_qmap = pd.concat([drq_qmap, tvq_qmap])
question_map = combined_qmap.set_index('id')['question'].to_dict()
```

```python


import matplotlib.pyplot as plt
import seaborn as sns

# Calculate standard deviation per question to find the most divisive ones
question_variability = pivot_avg.std(axis=1)

top5_questions = question_variability.sort_values(ascending=False).head(5)

sns.set(style="whitegrid")

for i in range(5):
    question_id = top5_questions.index[i]
    plt.figure(figsize=(10, 5))
    data = pivot_avg.loc[question_id].sort_values()

    df_plot = pd.DataFrame({
        'Parti': data.index,
        'Average answer': data.values
    })

    sns.barplot(
        data=df_plot,
        x='Average answer',
        y='Parti',
        hue='Parti',          
        palette='coolwarm',
        dodge=False,
        legend=False        
    )

    full_question = question_map.get(str(question_id), "Unknown question")
    
    plt.title(f"Average answer by parti: {full_question}")
    plt.xlabel("Average answer svar")
    plt.ylabel("Parti")
    plt.xlim(-2, 2)
    plt.tight_layout()
    plt.show()
```

### Clustering Analysis
### Prep data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, ward
from sklearn.cluster import AgglomerativeClustering

alldata = pd.read_excel('alldata.xlsx')
electeddata = pd.read_excel('electeddata.xlsx')

plt.rcParams['figure.figsize'] = (12, 10)
party_colors = {
    'Socialdemokratiet': '#b32722', 'Radikale Venstre': '#713280', 'Konservative': '#9eba27',
    'Nye Borgerlige': '#054f5c', 'Socialistisk Folkeparti': '#dd0878', 'Liberal Alliance': '#21c6cf',
    'Kristendemokraterne': '#bababa', 'Moderaterne': '#6c39e5', 'Dansk Folkeparti': '#f3cd41',
    'Frie Grønne': '#00e66a', 'Venstre': '#006697', 'Danmarksdemokraterne': '#8ac3ff',
    'Enhedslisten': '#ed8019', 'Alternativet': '#2a8f39',

}

dr_questions = [str(col) for col in range(530, 564)]
tv_questions = [f"{i}{l}" for i in range(1, 13) for l in ['a', 'b']]
question_columns = [col for col in dr_questions + tv_questions if col in alldata.columns]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(alldata[question_columns])

pca = PCA(n_components=2, random_state=13)
X_pca = pca.fit_transform(X_scaled)
alldata['PCA1'], alldata['PCA2'] = X_pca[:,0], X_pca[:,1]
```

### K-Means

```python
n_parties = alldata['parti'].nunique()
kmeans = KMeans(n_clusters=n_parties, random_state=13, n_init=10)
alldata['cluster_labels'] = kmeans.fit_predict(X_scaled)

# K-Means
plt.figure()
for cl in sorted(alldata['cluster_labels'].unique()):
    subset = alldata[alldata['cluster_labels'] == cl]
    plt.scatter(subset['PCA1'], subset['PCA2'],
                label=f'Cluster {cl}', alpha=0.6, s=40)
plt.title(f'KMeans Clustering (k={n_parties})')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(bbox_to_anchor=(1,1), loc='upper left')
plt.show()

# actual party clusterization
plt.figure()
for party in alldata['parti'].unique():
    subset = alldata[alldata['parti'] == party]
    plt.scatter(subset['PCA1'], subset['PCA2'],
                label=party,
                color=party_colors.get(party, 'grey'),
                alpha=0.6, s=40)
plt.title('PCA Projection — Colored by Party')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(bbox_to_anchor=(1,1), loc='upper left')
plt.show()
```

### DBSCAN

```python
dbscan = DBSCAN(eps=3.5, min_samples=3)
alldata['DBSCAN_Cluster'] = dbscan.fit_predict(X_scaled)

plt.figure()
for label in sorted(alldata['DBSCAN_Cluster'].unique()):
    subset = alldata[alldata['DBSCAN_Cluster'] == label]
    color = 'grey' if label == -1 else None
    plt.scatter(subset['PCA1'], subset['PCA2'], label=f'Cluster {label}',
                alpha=0.6, s=40, c=color)

plt.title('DBSCAN Clustering (eps=3.5, min_samples=3)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

### Hierarchical clustering

```python

# dendrogram
linked = ward(X_pca)

plt.figure(figsize=(14, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=False)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()

# visualisation of clusters
agg_clustering = AgglomerativeClustering(n_clusters=n_parties, linkage='ward')
agg_labels = agg_clustering.fit_predict(X_pca)

alldata['agglomerative_cluster'] = agg_labels

plt.figure()
for cluster in sorted(alldata['agglomerative_cluster'].unique()):
    subset = alldata[alldata['agglomerative_cluster'] == cluster]
    plt.scatter(subset['PCA1'], subset['PCA2'],
                label=f'Cluster {cluster}',
                alpha=0.7, s=50)
plt.title(f'Agglomerative Clustering on PCA Components (k={n_parties})')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

### Political Compass

```python
economic_questions = [
    '10a', '10b', '11a', '12a', '12b', '1b', '4a', '531', '534', '535', '537', '540',
    '541', '543', '545', '548', '550', '551', '555', '556', '557', '5a', '6a', '6b',
    '7a', '8a'
]

social_questions = [
    '11b', '1a', '2a', '2b', '3a', '3b', '4b', '530', '533', '538', '544', '546',
    '547', '552', '553', '559', '561', '563', '5b', '7b', '8b', '9a', '9b'
]
# The categorization of questions was done manually with the help of ChatGPT.
# With more data and labeled questions (e.g. economic vs social), we could train a model to categorize them automatically.


alldata['economic_score'] = alldata[economic_questions].mean(axis=1)
alldata['social_score'] = alldata[social_questions].mean(axis=1)

party_means = alldata.groupby('parti')[['economic_score', 'social_score']].mean().reset_index()

plt.figure()
# Individual candidates
for party in alldata['parti'].unique():
    party_data = alldata[alldata['parti'] == party]
    plt.scatter(party_data['economic_score'], party_data['social_score'],
                alpha=0.2, color=party_colors.get(party, 'grey'), s=20)

# Party averages
for party in party_means['parti']:
    row = party_means[party_means['parti'] == party]
    x, y = row['economic_score'].values[0], row['social_score'].values[0]
    plt.scatter(x, y, color=party_colors.get(party, 'grey'), s=150, edgecolors='black', label=party)

plt.legend(bbox_to_anchor=(1, 1))

plt.text(1.15, 1.15, 'Progressive Right', fontsize=11, alpha=0.6, ha='right', va='top')
plt.text(-1.15, 1.15, 'Progressive Left', fontsize=11, alpha=0.6, ha='left', va='top')
plt.text(1.15, -1.15, 'Conservative Right', fontsize=11, alpha=0.6, ha='right', va='bottom')
plt.text(-1.15, -1.15, 'Conservative Left', fontsize=11, alpha=0.6, ha='left', va='bottom')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.xlabel('Economic Axis')
plt.ylabel('Social Axis')
plt.title('Political Compass')
plt.grid(True, alpha=0.3)
plt.show()
```

### Party internal agreement

```python
pca_elected = alldata[alldata['navn'].isin(electeddata['navn'])]
results = []
for party, group in pca_elected.groupby('parti'):
    coords = group[['PCA1', 'PCA2']].values
    if len(coords) < 2:
        continue
    dists = np.linalg.norm(coords[:, None] - coords, axis=2)
    avg_dist = -np.mean(dists[np.triu_indices(len(coords), k=1)])
    results.append((party, avg_dist, len(coords)))

df_agree = pd.DataFrame(results, columns=['Party', 'Avg_Internal_Agreement', 'Count'])
df_agree = df_agree.sort_values('Avg_Internal_Agreement')

plt.figure()
plt.barh(df_agree['Party'], df_agree['Avg_Internal_Agreement'],
         color=[party_colors.get(p, 'grey') for p in df_agree['Party']])
plt.xlabel('Average Internal Agreement (lower distance = more similar)')
plt.title('Agreement Within Each Party')
plt.tight_layout()
plt.show()
```
