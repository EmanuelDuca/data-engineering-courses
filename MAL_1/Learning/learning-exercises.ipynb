```python
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis

data = pd.read_csv('titanic.csv').dropna()
# We will take the Age as target group
age_mean = data['Age'].mean()
age_median = data['Age'].median()
age_mode = data['Age'].mode()
print('_Age_\nmean:',age_mean,'\nmedian:',age_median,'\nMode:',age_mode)

# Standerd Deviation
age_std = data['Age'].std()
age_var = data['Age'].var()

print('Std:',age_std,"Var:",age_var)

# Compute specific quantiles
quantiles = data['Age'].quantile([0.25, 0.5, 0.75])
print(quantiles)

ages = data['Age'].dropna()

age_skew = skew(ages)
age_kurt = kurtosis(ages)

print('Skewness:', age_skew)
print('Kurtosis:',age_kurt)

```

```python
# For creating a plot 
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(ages,bins=
             40,kde=True)
plt.show()
```
