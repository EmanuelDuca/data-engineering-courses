##### Integrate in Py

```python
import scipy.integrate as integrate
import sympy as sp 

x = sp.symbols('x')
f = 1 / (1 + x)**2

# Compute the indefinite integral
integral = sp.integrate(f,x)
print(integral)

# Define integral from 0 to 1
definite_integral = sp.integrate(f,(x,0,1))
print(definite_integral)

```

##### Double Integrals

```python
import scipy.integrate as integrate
import sympy as sp

# Define symbols
x, y = sp.symbols('x y')

# Define the function
f = (1/2) * sp.E**(-x) + y / (1 + x)**2


# Define limits
x_lower = 0
x_upper = 1
y_lower = 0
y_upper = 1/2

# Step 1: Definite integral over x from 0 to 1
integral_x = sp.integrate(f, (x, x_lower, x_upper))
print("Definite integral over x (0 to 1):")
print(integral_x)

# Step 2: Definite integral over y from 0 to 1/2
integral_xy = sp.integrate(integral_x, (y, y_lower, y_upper))
print("\nDefinite double integral over y (0 to 1/2):")
print(integral_xy)

numeric_result = integral_xy.evalf()
print('\nNumberic result: ',numeric_result)

```

```python
import scipy.integrate as integrate
import sympy as sp

# Define symbols
x, y = sp.symbols('x y')

# Define the function
f = sp.E**(-x) * (3/8) * y**2

# Define limits
x_lower = 0
x_upper = sp.oo
y_lower = 0
y_upper = 1/2

# Marginal of Y: integrate out x from 0 to ∞
f_Y_integral = sp.integrate(f,(x,x_lower, x_upper))
print(f_Y_integral)

# FInd out the Variance of Y 

f_Y = (3/8) * y**2

# Define the limits of integration
lower_limit = 0
upper_limit = 2

# Step 1: Compute E[Y]
E_Y = sp.integrate(y * f_Y, (y, lower_limit, upper_limit))
print("E[Y] =", E_Y)

# Step 2: Compute E[Y^2]
E_Y2 = sp.integrate(y**2 * f_Y, (y, lower_limit, upper_limit))
print("E[Y^2] =", E_Y2)

# Step 3: Compute Var(Y)
Var_Y = E_Y2 - E_Y**2
print("Var(Y) =", Var_Y)



```
