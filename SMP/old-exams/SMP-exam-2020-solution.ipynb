```python
# b)
n10 = len(df['2010s models'])
m10 = np.mean(df['2010s models'])
SE10 = stats.sem(df['2010s models'])
Level = 0.95

CI10 = stats.t.interval(Level, n10-1, loc=m10, scale=SE10)

n20 = len(df['2020s models'])
m20 = np.mean(df['2020s models'])
SE20 = stats.sem(df['2020s models'])

CI20 = stats.t.interval(Level, n20-1, loc=m20, scale=SE20)

display(Latex('A ' + repr(int(Level*100)) + ' % confidence interval for the sample mean of the 2010 models is ['+
              repr(round(CI10[0],3)) + ' ; ' + repr(round(CI10[1],3)) + ']'))
display(Latex('A ' + repr(int(Level*100)) + ' % confidence interval for the sample mean of the 2020 models is ['+
              repr(round(CI20[0],3)) + ' ; ' + repr(round(CI20[1],3)) + ']'))


# Confidence interval calculations
CI10 = stats.norm.interval(0.95, loc=m10, scale=SE10)
CI20 = stats.norm.interval(0.95, loc=m20, scale=SE20)

# Plotting
plt.figure(figsize=(10, 6))
x10 = np.linspace(m10 - 4*SE10, m10 + 4*SE10, 1000)
z10 = np.linspace(CI10[0], CI10[1], 1000)
y10 = stats.norm.pdf(x10, m10, SE10)
plt.plot(x10, y10, color='blue', label='2010s models PDF')
plt.fill_between(z10, stats.norm.pdf(z10, m10, SE10), color='blue', alpha=0.5, label='2010s CI')

x20 = np.linspace(m20 - 4*SE20, m20 + 4*SE20, 1000)
z20 = np.linspace(CI20[0], CI20[1], 1000)
y20 = stats.norm.pdf(x20, m20, SE20)
plt.plot(x20, y20, color='green', label='2020s models PDF')
plt.fill_between(z20, stats.norm.pdf(z20, m20, SE20), color='green', alpha=0.5, label='2020s CI')

plt.title('Probability Density Function and Confidence Intervals for 2010s and 2020s Models')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# c)
# Prepare data
data = df['2010s models']
skewness = round(stats.skew(data), 4)
kurtosis = round(stats.kurtosis(data), 4)
kurtosis_test = stats.kurtosistest(data)

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Probability plot on the first subplot
stats.probplot(data, plot=ax1)
ax1.set_title('Probability Plot of 2010s Models')
ax1.set_ylabel('2010s Models')
ax1.grid(True)

# Kernel density estimate on the second subplot
data.plot.kde(ax=ax2, legend=False, title='KDE of 2010s Models')
ax2.grid(True)

# Add annotations for skewness and kurtosis
textstr = f'Skewness: {skewness}\nKurtosis: {kurtosis}\nKurtosis Test p-value: {kurtosis_test.pvalue:.4f}'
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
fig.text(0.5, -0.05, textstr, fontsize=12, ha='center', va='center', bbox=props)

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Prepare data
data = df['2020s models']
skewness = round(stats.skew(data), 4)
kurtosis = round(stats.kurtosis(data), 4)
kurtosis_test = stats.kurtosistest(data)

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Probability plot on the first subplot
stats.probplot(data, plot=ax1)
ax1.set_title('Probability Plot of 2020s Models')
ax1.set_ylabel('2020s Models')
ax1.grid(True)

# Kernel density estimate on the second subplot
data.plot.kde(ax=ax2, legend=False, title='KDE of 2020s Models')
ax2.grid(True)

# Add annotations for skewness and kurtosis
textstr = f'Skewness: {skewness}\nKurtosis: {kurtosis}\nKurtosis Test p-value: {kurtosis_test.pvalue:.4f}'
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
fig.text(0.5, -0.05, textstr, fontsize=12, ha='center', va='center', bbox=props)

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# As can be seen, it is fair to assume that both datasets are normally distributed. The normal probability plots
# show distributions that almost perfectly follow the trendline. The distribution plot of the 2010 model is also
# perfectly bell-shaped, whereas the one for the 2020 models is acceptable so.

# If skewness is between -0.5 and 0.5, the distribution is approximately symmetric which is the case for both datasets.
# The KurtosisTest tests the null hypothesis that the kurtosis of the population from which the sample was drawn is
# that of the normal distribution. We see we get significant p-value which would indicate the data
# is not normally distributed.
```

```python
# d)
from matplotlib.ticker import FuncFormatter 

# Calculate the difference and standard error
diff = abs(m10 - m20)
std = np.sqrt(df['2010s models'].var(ddof=1)/n10 + df['2020s models'].var(ddof=1)/n20)
Level = 0.99

# Calculate the confidence interval
CIdiff = stats.t.interval(Level, df=(n10+n20-2), loc=diff, scale=std)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
xdiff = np.linspace(diff-4*std, diff+4*std, 1000)
zdiff = np.linspace(CIdiff[0], CIdiff[1], 1000)
ydiff = stats.t.pdf(xdiff, n10+n20-2, diff, std)
ax.plot(xdiff, ydiff, color='dodgerblue', label='Probability Density Function')
ax.fill_between(zdiff, stats.t.pdf(zdiff, n10+n20-2, diff, std), color='dodgerblue', alpha=0.4, label='99% Confidence Interval')

# Enhance plot with labels and title
ax.set_title('Confidence Interval for Mean Difference of Two Models')
ax.set_xlabel('Mean Difference')
ax.set_ylabel('Probability Density')
ax.legend()

# Format the x-axis to show the confidence interval
def formatter(x, pos):
    if x in CIdiff:
        return f'{x:.3f}'
    return ''

ax.xaxis.set_major_formatter(FuncFormatter(formatter))
plt.grid(True)
plt.show()

# Displaying the confidence interval as text on the plot
confidence_text = f'99% Confidence Interval: [{CIdiff[0]:.3f}, {CIdiff[1]:.3f}]'
ax.text(0.5, -0.1, confidence_text, transform=ax.transAxes, fontsize=12, ha='center', color='dodgerblue')
```

```python
# e)
# H0: Mean difference is 0
# H1: Mean difference is not 0
alpha = 0.05
p = stats.t.cdf(diff, df=n10+n20-2, loc=0, scale=std)
pvalue = 2*min(p, 1-p)

if round(pvalue, 4) < alpha:
    print("Reject since p-value = " + repr(round(pvalue, 4)) + ' < ' + repr(alpha))
else:
    print("Fail to reject since p-value = " + repr(round(pvalue, 4)) + ' > ' + repr(alpha))
```

```python
from IPython.display import Markdown

# Load data
df = pd.read_excel("Wages_and_Work_Hours.xlsx")

# Creating a contingency table
table = pd.crosstab(df['Gender'], df['Education Level'])
display(Markdown("### Contingency Table"))
display(table)

# Define hypotheses
# H0: Gender is independent of Education Level.
# H1: Gender is dependent on Education Level.


# Perform Chi-squared test for independence
alpha = 0.05
Chi2, pvalue, dof, expected = stats.chi2_contingency(table, correction=False)

# Calculate the critical value from the chi-square distribution
critical_value = stats.chi2.ppf(1 - alpha, dof)

# Display the critical value
display(Markdown(f"**Critical Value:** {critical_value:.4f}"))

# Display the test results
display(Markdown(f"**Chi-squared Statistic:** {Chi2:.4f}, **Degrees of Freedom:** {dof}"))
display(Markdown(f"**Expected Frequencies:**"))
display(pd.DataFrame(expected, index=table.index, columns=table.columns))

# Decision on the hypothesis test
if pvalue < alpha:
    result = f"Reject $H_0$ since p-value = {pvalue:.4f} < {alpha}"
else:
    result = f"Fail to reject $H_0$ since p-value = {pvalue:.4f} > {alpha}"
display(Markdown(f"**Test Conclusion:** {result}"))

```
