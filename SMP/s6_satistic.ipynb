##### Lesson 6 - Statistical Intervals
##### Two-sided confidence

```python
from scipy.stats import norm
import numpy as np

x_mean = 74.036
sigma = 0.001
n = 15
confidence = 99
alpha = 1-confidence/100

z_score = norm.ppf(1- alpha/2)
#z_score = 2.58
print(z_score)

margin_error = z_score * (sigma/np.sqrt(n))
margin_error

# Compute CI
ci_lower = x_mean - margin_error
ci_upper = x_mean + margin_error
print(f"{confidence}% Confidence Interval: [{ci_lower}, {ci_upper}]")

```

##### Lower-confidence

```python
z_score = norm.ppf(1-alpha)
z_score = 2.33
print(z_score)

margin_error = z_score * (sigma / np.sqrt(n))

# Compute lower one-sided CI
ci_lower = x_mean - margin_error

print(f"{confidence}% Lower Confidence Bound: [{ci_lower}, ∞)")
```

```python
def conver_seconds_to_year(number):
    secInYear = 365*24*60*60 
    years = number // secInYear
    remain = number %  secInYear
    
    months = remain // (secInYear /12)
    remain = remain % (secInYear/12)
    
    secInDay = 60*60*24
    days = remain // secInDay
    
    print("years:",years,"month:",months,"days:",days)

conver_seconds_to_year(1000000)
conver_seconds_to_year(1000000000)
conver_seconds_to_year(100000000000000)
        
```
