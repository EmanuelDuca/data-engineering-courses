```python
#Ex 2.2 
"""
120 Passangers. 125 sold tickes. Not showing up 0.1 probability. Passangers behave independently.
a) What is the probability that every passangers that shows up can take the filght ?
b)  What is the probability that the flight departs with empty seats. ?
"""

# A - A susccess is not showing up
n = 125 #tickets
p = 0.1 

# P(Y ≥ 5) = 1 - P(Y ≤ 4)
prob_all_seated = 1 - binom.cdf(4, n, p)
print("P(all passengers can take the flight) =", round(prob_all_seated, 4))

# P(Y ≥ 6) = 1 - P(Y ≤ 5)
prob_empty_seats = 1 - binom.cdf(5, n, p)
print("P(flight departs with empty seats) =", round(prob_empty_seats, 4))
```
