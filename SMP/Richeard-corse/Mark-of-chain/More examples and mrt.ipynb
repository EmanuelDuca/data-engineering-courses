```python
from sympy import *
from sympy import Matrix as M
from sympy import Rational as R
from IPython.display import display, Markdown
```

```python
# Example 7
a1, a2 = symbols('a1 a2')
a0 = 0
a3 = 1
a = [a0, a1, a2, a3]
P = [[1,0,0,0],[R(1,3),0,R(2,3),0],[0,R(1,2),0,R(1,2)],[0,0,0,1]]

# Calculate the sum for each a_i
sum_results = [sum(a[k] * P[i][k] for k in range(len(a))) for i in range(len(a))]

# Print the results
for i, result in enumerate(sum_results):
    # Check if the result is a number or a sympy expression
    if isinstance(result, sp.Expr):  # if it's a SymPy expression
        # Use SymPy's display function to show expressions
        display(sp.Eq(sp.Symbol(f'a_{i}'), result))
    else:
        # If it's just a number, convert to float and format
        display(Markdown(f"$a_{{{i}}} = {float(result):.2f}$"))
```

```python
# Example: Lipstick

P = M([[0.7, 0.2, 0.1], [0.5, 0.4, 0.1], [0.6, 0.2, 0.2]])

# Step 1
X = P.T

# Step 2
X = X - eye(P.shape[0])

# Step 3
X = X[0:-1,:]

# Step 4
X = X.row_join(zeros(P.shape[0]-1, 1))

# Step 5
X = X.col_join(ones(1,P.shape[1]+1))
X.rref()[0]
```
